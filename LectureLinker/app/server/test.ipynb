```python
import requests
def get_stream(query: str):
    s = requests.Session()
    with s.get(
        "http://localhost:8000/chat/${query}",
        stream=True,
    ) as r:
        for line in r.iter_content():
            print(line.decode("iso-8859-1"), end="")
get_stream("Klausur")
```

```python
import requests

def get_stream(query: str):
    s = requests.Session()
    with s.post(
        "http://localhost:8000/chat",
        stream=True,
        json={"text": query}
    ) as r:
        for line in r.iter_content():
            print(line.decode("utf-8"), end="")

get_stream("hi there!")
```
